# Fundamentos tericos

La Regresin Lineal es una tcnica estadstica que busca modelar la relacin entre dos variables, una variable dependiente (Y) y una o ms variables independientes (X), mediante una lnea recta que mejor se ajuste a los datos observados. Es decir, se utiliza para predecir o estimar el valor de una variable dependiente basndose en los valores de una o ms variables independientes.

La ecuacin de regresin lineal se representa matemticamente como:

$$ Y=\beta_0+\beta_iX+\in_i $$

Donde:

$Y$ es la variable dependiente que se desea predecir o estimar.
$X$ es la variable independiente o variables independientes que se utilizan para hacer la prediccin o estimacin.
$\beta_0$ es el intercepto, que representa el valor de $Y$ cuando $X$ es igual a cero.
$\beta_1$ es la pendiente, que representa el cambio en $Y$ por unidad de cambio en $X$.
$\epsilon$ es el error, que representa la diferencia entre el valor observado de $Y$ y el valor predicho o estimado de $Y$.


## Regresin Lineal Simple:

La regresin lineal simple es un mtodo estadstico utilizado para modelar la relacin lineal entre dos variables, una variable independiente (denotada como $X$) y una variable dependiente (denotada como $Y$). La ecuacin de regresin lineal simple se representa matemticamente de la siguiente manera:
La ecuacin de regresin lineal simple es:

$$ Y=\beta_0+\sum\;\beta_iX+\in_i $$

Donde:

$Y$ es la variable dependiente que se desea predecir.
$X$ es la variable independiente que se utiliza para hacer la prediccin.
$\beta_0$ es el intercepto, que representa el valor de $Y$ cuando $X$ es igual a cero.
$\beta_1$ es la pendiente, que representa el cambio en $Y$ por unidad de cambio en $X$.
$\epsilon$ es el error, que representa la diferencia entre el valor observado de $Y$ y el valor predicho de $Y$.


## Regresin Lineal Mltiple:
La regresin lineal mltiple es un mtodo utilizado para modelar la relacin lineal entre una variable dependiente y dos o ms variables independientes. La ecuacin de regresin lineal mltiple se representa matemticamente de la siguiente manera:
La ecuacin de regresin lineal mltiple es:

$$ Y=\beta_0+\;\beta_1X_1+\beta_2X_2+...+\beta_pX_p+\in $$

Donde:

$Y$ es la variable dependiente que se desea predecir.
$X_1, X_2, \ldots, X_p$ son las variables independientes que se utilizan para hacer la prediccin.
$\beta_0$ es el intercepto, que representa el valor de $Y$ cuando todas las variables independientes son iguales a cero.
$\beta_1, \beta_2, \ldots, \beta_p$ son las pendientes, que representan el cambio en $Y$ por unidad de cambio en cada una de las variables independientes.
$\epsilon$ es el error, que representa la diferencia entre el valor observado de $Y$ y el valor predicho de $Y$.





# Desarrollo del ejercicio

## Paso1

Importar las librerias requeridas



```python
import matplotlib.pyplot as plt
import numpy as np
```

## Paso 2

Crear la funcin



```python
# Crear la funcin regresin lineal
def regress(x, y, n):    
    sumx = 0                 # Sumatoria de la variable independiente
    sumxy = 0                # Sumatoria del producto entre x, y
    st = 0                   # Suma total de los cuadrados de las diferencias en los datos
    sumy = 0                 # Sumatoria de y que es la variable dependiente
    sumx2 = 0                # Suma de los valores de x elevados al cuadrado
    sr = 0                   # Suma residuales al cuadrado

    for i in range(n):     
        sumx = sumx + x[i]    
        sumy = sumy + y[i]    
        sumxy = sumxy + x[i]*y[i] 
        sumx2 = sumx2 + x[i]*x[i] 
  
    promx = sumx/n                # Promedio de los valores en x                          
    promy = sumy/n                # Promedio de los valores en y                          
    b = (n*sumxy-sumx*sumy)/(n*sumx2-sumx*sumx)  # Coeficiente de regresin (pendiente)
    a = np.mean(y)-b*np.mean(x)                  # Intercepto de regresin

    for i in range(n):
        st = st + (y[i]-promy)**2
        sr = sr + (y[i]-b*x[i]-a)**2
  
    syx = (sr/(n-2))**0.5         # Error estndar de la estimacin
    r2 = (st-sr)/st              # Coeficiente de determinacin (R2)

    return b, a, syx, r2
```

## Paso 3

Ingreso de los datos


```python
n = 9                                       # Nmero de datos en los arreglos x e y
x = [2, 4, 6, 7, 10, 11, 14, 17, 20]        # Datos de la variable independiente x
y = [2, 1, 7, 3, 8, 9, 5, 8, 14]            # Datos de la variable dependiente y
```

## Paso 4

Imprimir los datos solicitados


```python
b, a, syx, r2 = regress(x, y, len(x))        # Llamada a la funcin regress() con los arreglos x e y y su longitud n
print("Pendiente: ", b)                       # Imprimir el valor de la pendiente (coeficiente de regresin)
print("Interseccin: ", a)                    # Imprimir el valor de la interseccin (intercepto de regresin)
print("Error Estandar: ", syx)                 # Imprimir el valor del error estndar de la estimacin
print("Coeficiente de correlacin: ", r2)   # Imprimir el valor del coeficiente de correlacin (R2)
```

## Paso 5

Crear la linea que se ajusta a los datos

```python
t = np.poly1d ([b, a]) # y = bx + a
print(t)
```

## Paso 6 Crear la nube de puntos



```python
x = np.array([2, 4, 6, 7, 10, 11, 14, 17, 20])
y = np.array([2, 1, 7, 3, 8, 9, 5, 8, 14])
plt.plot(x,y,'*', label='Datos')
plt.xlabel('Valores x')
plt.ylabel('Valores y')
plt.title('Nube de puntos')
plt.grid()
plt.show()
```

## Paso 7 Creacin de la grfica

```python
xi = np.array([2, 4, 6, 7, 10, 11, 14, 17, 20])
yi = np.array([2, 1, 7, 3, 8, 9, 5, 8, 14])
plt.plot(xi, yi, '*', label='Datos')
plt.plot(x, b * x + a, label='Ajuste')
plt.xlabel('Valores x')
plt.ylabel('Valores y')
plt.title('Ajuste de curva')
plt.grid()
plt.legend (loc=4)
plt.show()
```

El error estndar de la estimacin (syx) proporciona una medida de la precisin de la regresin, siendo menor su valor, mejor ser la calidad del ajuste.
